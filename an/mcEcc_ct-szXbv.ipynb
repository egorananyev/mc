# Behavioural analysis

## Preparing the data

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os, glob
```

Loading the data:

```python
%matplotlib inline
plt.style.use('dark_background')
expName = 'mcEcc_ct-szXbv'
dataDir = os.path.abspath(os.path.join(os.getcwd(), '..', '..', 'data/', expName))
allSubjDirs = glob.glob(dataDir + '/' + expName + '*')
#print allSubjDirs
ds = pd.DataFrame()
for subjDir in allSubjDirs:
    print subjDir
    subjFilePath = glob.glob(subjDir + '/' + expName + '*.csv')[0]
    #print subjFilePath
    #ss = pd.read_csv(subjFilePath)
    ds = ds.append(pd.read_csv(subjFilePath))
#print ds.head(5)
print ds.shape
print ds.columns
```

Transforming the data set first.

```python
ds['stimSz'] = np.round((ds.periGap*2+32)/35,1)
ds['ringSzDeg'] = np.round((ds.ringSz+32)/35,1)
ds['ringSzP'] = np.round(ds.ringSzDeg/ds.stimSz,2)
ds = ds.reset_index(drop=True) # resetting the index as there are row duplicates
print ds.head(5)
```

## Visualization

```python
summary = ds.groupby(['participant'], as_index=False)['nf270'].agg([np.sum, np.mean, np.std, np.size]).reset_index()
print 'Slow (down):'
print summary
summary = ds.groupby(['participant'], as_index=False)['nf180'].agg([np.sum, np.mean, np.std, np.size]).reset_index()
print 'Fast (left):'
print summary
summary = ds.groupby(['participant'], as_index=False)['nf090'].agg([np.sum, np.mean, np.std, np.size]).reset_index()
print 'Patchy/transparent (up):'
print summary
```

```python
ssRing = ds[ds.ringSzP < 1]
```

### Proportion slow/fast/patchy trials

```python
g = sns.factorplot(x='stimSz', y='nf270', col='participant', data=ds, ci=95, dodge=.0, size=2.5)
plt.ylim(-0.1,1.1)
g = sns.factorplot(x='stimSz', y='nf180', col='participant', data=ds, ci=95, dodge=.0, size=2.5)
plt.ylim(-0.1,1.1)
g = sns.factorplot(x='stimSz', y='nf090', col='participant', data=ds, ci=95, dodge=.0, size=2.5)
plt.ylim(-0.1,1.1)
```

```python
summary = ds.groupby(['participant','stimSz'], as_index=False)['nf270','nf180','nf090'].agg([np.mean]).reset_index()
sumPtnts = pd.DataFrame()
sumPtnts['participant'] = summary.participant
sumPtnts['stimSz'] = summary.stimSz
sumPtnts['static'] = summary.nf270['mean']
sumPtnts['dynamic'] = summary.nf180['mean']
sumPtnts['patchy'] = summary.nf090['mean']
sumPtntsMelted = pd.melt(sumPtnts, id_vars=['participant','stimSz'])
sumStimSz = sumPtntsMelted.rename(columns={'variable': 'response', 'value': 'proportion trials'})
sumStimSz.head(5)
```

```python
g = sns.factorplot(x='stimSz', y='proportion trials', hue='response', col='participant', data=sumStimSz, 
               col_wrap=4, size=3.5, ci=None)
g.set_axis_labels('stimSz (deg)','Proportion trials')
plt.ylim(-0.1,1.1)
```
